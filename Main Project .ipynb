Pre-processing
---------------


```python
#missing values
df.isnull().sum()
```

```python
#remove missing values
df = df.dropna()
```

```python
#Removing Duplicates
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()
```

```python
#Correcting Data type

df.dtypes
```

```python
categorical_cols = [
    'Gender', 'Education_Level', 'Employment_Status',
    'Household_Income', 'Location_Type',
    'Engagement_Level', 'Employment_Impact'
]

df[categorical_cols] = df[categorical_cols].astype('category')
```

```python
#Creating Derived column

df['Computer_Knowledge_Improvement'] = (
    df['Post_Training_Basic_Computer_Knowledge_Score'] -
    df['Basic_Computer_Knowledge_Score']
)

df['Internet_Usage_Improvement'] = (
    df['Post_Training_Internet_Usage_Score'] -
    df['Internet_Usage_Score']
)

df['Mobile_Literacy_Improvement'] = (
    df['Post_Training_Mobile_Literacy_Score'] -
    df['Mobile_Literacy_Score']
)
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 25, 40, 60, 100],
    labels=['Youth', 'Adult', 'Middle_Aged', 'Senior']
)
```

```python
#Filtering or aggregating data
df = df[df['Age'] > 18]
```

```python
avg_literacy_by_location = (
    df.groupby('Location_Type', observed=False)['Overall_Literacy_Score']
      .mean()
      .reset_index()
)

avg_literacy_by_location
```

Visualisation
-------------

Gender Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Main_Project.csv")
```

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Shows participation across Male, Female, and Other

Age Distribution

```python
plt.figure()
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Displays how users are spread across age groups

Overall Literacy Score by Gender

```python
plt.figure()
df.boxplot(column='Overall_Literacy_Score', by='Gender')
plt.title("Overall Literacy Score by Gender")
plt.suptitle("")
plt.xlabel("Gender")
plt.ylabel("Overall Literacy Score")
plt.show()
```

This compares literacy score distribution across genders

Internet Usage vs Overall Literacy Score

```python
plt.figure()
plt.scatter(df['Internet_Usage_Score'], df['Overall_Literacy_Score'])
plt.title("Internet Usage vs Overall Literacy Score")
plt.xlabel("Internet Usage Score")
plt.ylabel("Overall Literacy Score")
plt.show()
```

This examines relationship between two numerical variables

Employment Impact

```python
plt.figure()
df['Employment_Impact'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Employment Impact")
plt.ylabel("")
plt.show()
```

This shows proportion of users who experienced employment impact

Correlation Matrix

```python
plt.figure()
corr = df.select_dtypes(include=np.number).corr()
plt.imshow(corr)
plt.title("Correlation Heatmap")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.show()
```

Here it displays correlation between numerical variables

Average Overall Literacy Score by Age

```python

age_literacy = df.groupby('Age')['Overall_Literacy_Score'].mean()

plt.figure()
plt.plot(age_literacy.index, age_literacy.values)
plt.title("Average Overall Literacy Score by Age")
plt.xlabel("Age")
plt.ylabel("Average Literacy Score")
plt.show()
```

Here it shows how literacy score changes across age groups
